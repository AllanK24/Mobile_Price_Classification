```python
import pandas as pd
```

```python
#Load the dataset
X = pd.read_csv(r"C:\Users\Windows 10\Desktop\mobile_price_classification\dataset\train.csv")
y = X["price_range"]
X = X.drop(["price_range"],axis=1)
```

```python
#print out the information about the dataset
X.info()
y.info()
```

```python
#taking a look at the features
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
"""
Applying Feature scaling
Features to be scaled:
    --battery_power
    --px_height
    --px_width
    --ram
"""
#Choosing features for scaling
scale = X[["battery_power","px_height","px_width","ram"]].copy()
print(scale.describe())

#Scale features
scaler = StandardScaler()
scaler.fit_transform(scale)
scale = scaler.transform(scale)
print(scale)
```

```python
#Replacing with the scaled data
X["battery_power"] = scale[:,0]
X["px_height"] = scale[:,1]
X["px_width"] = scale[:,2]
X["ram"] = scale[:,3]

print("After scaling:")
X.describe()
```

```python
#Splitting into training and validation sets
X_train = X[:1800]
X_val = X[1800:]
y_train = y[:1800]
y_val = y[1800:]
```

```python
#Building a Neural Network
import tensorflow as tf
from keras import Sequential
from keras.layers import Dense
from keras.regularizers import l2

model = Sequential([
    Dense(units=20,activation="relu",kernel_regularizer=l2(0.01)),
    Dense(units=12,activation="relu",kernel_regularizer=l2(0.01)),
    Dense(units=4,activation="softmax")
])

#loss cost and optimizer
from keras.losses import SparseCategoricalCrossentropy
from keras.optimizers import Adam
from keras.metrics import SparseCategoricalAccuracy

model.compile(optimizer=Adam(0.001),loss=SparseCategoricalCrossentropy(),metrics=[SparseCategoricalAccuracy()])
model.fit(X_train,y_train,epochs=150,batch_size=64)
```

```python
#Inference on validation set
y_inf = model.predict(X_val)
```

```python
#Getting maximum probability from predictions
import numpy as np

counter = 0
row = np.zeros(4)
y_hat = np.zeros(len(y_val))
while counter<len(y_val):
    row = y_inf[counter]
    #y_hat[counter] = np.max(row)
    max_num = np.max(row)
    max_in = 0
    for prob in row:
        if max_num==prob:
            break
        max_in+=1
    y_hat[counter] = max_in
    counter+=1
```

```python
#Converting pandas y_val to numpy y_val
y_val = y_val.to_numpy()
```

```python
#Measuring accuracy
counter = 0
correct = 0
equal = False
while counter<len(y_val):
    y_hat[counter] = int(y_hat[counter])
    equal = y_hat[counter]==y_val[counter]
    if equal:
        correct+=1
    counter+=1
```

```python
#Accuracy on val set
print(f"Accuracy on validation set: {(correct/len(y_val))*100}%")
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_val,y_hat)
print(f"Confusion matrix: \n{cm}")
```
